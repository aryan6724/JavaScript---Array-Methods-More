# Question 5 : Explain how the sort method works in JavaScript and write code to sort an array of employee objects by their salary in descending order (highest salary first).
#// Array of employee objects
# const employees = [
# { name: "Rajesh", salary: 45000 },
# { name: "Priya", salary: 55000 },
# { name: "Amit", salary: 40000 },
# { name: "Kavya", salary: 60000 }];

1. Explanation: How the sort() Method Works in JavaScript

The sort() method is used to arrange (sort) elements of an array.

By default, it sorts values as strings, which can cause incorrect results for numbers.

To correctly sort numbers or objects, we provide a comparison function.

```
array.sort((a, b) => {
    return a - b;  // for ascending
});
```
2. Code to sort employees by salary in descending order

We want highest salary first, so we sort in descending:

```
// Array of employee objects
const employees = [
  { name: "Rajesh", salary: 45000 },
  { name: "Priya", salary: 55000 },
  { name: "Amit", salary: 40000 },
  { name: "Kavya", salary: 60000 }
];

// Sort by salary in descending order
const sortedEmployees = employees.sort((a, b) => b.salary - a.salary);

console.log(sortedEmployees);
```

// Shopping cart with items
const cart = [
# { name: "Laptop", price: 50000, quantity: 1 },
# { name: "Mouse", price: 500, quantity: 2 },
# { name: "Keyboard", price: 1500, quantity: 1 },
# { name: "Monitor", price: 15000, quantity: 1 }];

1. Total value of all items (price Ã— quantity)
```
// Shopping cart with items
const cart = [
  { name: "Laptop", price: 50000, quantity: 1 },
  { name: "Mouse", price: 500, quantity: 2 },
  { name: "Keyboard", price: 1500, quantity: 1 },
  { name: "Monitor", price: 15000, quantity: 1 }
];

// Total cart value
const totalValue = cart.reduce((total, item) => {
  return total + item.price * item.quantity;
}, 0);

console.log("Total Cart Value:", totalValue);
```

2. Find the most expensive single item (based on price)

```
const mostExpensive = cart.reduce((maxItem, item) => {
  return item.price > maxItem.price ? item : maxItem;
});

console.log("Most Expensive Item:", mostExpensive);
```

# Question 7 : Explain what array destructuring is and write code that extracts the first two colors from an array of colors into separate variables using destructuring.
# // Array of colors
# const colors = ["red", "blue", "green", "yellow", "purple"];

 - What is Array Destructuring?

Array destructuring is a feature in JavaScript that allows you to extract values from an array and store them into separate variables easily.
Instead of accessing values using indexes, destructuring makes the code shorter, cleaner, and more readable.

```
// Array of colors
const colors = ["red", "blue", "green", "yellow", "purple"];

// Destructuring the first two colors
const [firstColor, secondColor] = colors;

console.log(firstColor);   // Output: red
console.log(secondColor);  // Output: blue
```

# Question 8 : You have a student object with name, age, and marks properties. Write code using object destructuring to extract these values into separate variables and also rename the marks property to score while destructuring.
#// Student object
# const student = { name: "Arjun", age: 20, marks: 85, city: "Mumbai"};

```
// Student object
const student = {
  name: "Arjun",
  age: 20,
  marks: 85,
  city: "Mumbai"
};

// Object destructuring with renaming 'marks' to 'score'
const { name, age, marks: score } = student;

console.log(name);  // Output: Arjun
console.log(age);   // Output: 20
console.log(score); // Output: 85
```


# Question 9 : Write a function called calculateTotal that can accept any number of prices as arguments and returns their sum. Use the rest operator to handle the variable number of parameters.

- The rest operator lets a function accept any number of arguments and stores them in an array.

```
function calculateTotal(...prices) {
  let total = 0;
  
  prices.forEach(price => {
    total += price;
  });

  return total;
}

// Example call
console.log(calculateTotal(100, 200, 50));  // Output: 350
console.log(calculateTotal(10, 20));        // Output: 30
console.log(calculateTotal());              // Output: 0
```


# Question 10 : You have two arrays of student names and need to create a new array that contains all students from both arrays. Also, you have two objects representing a student's basic info and marks. Write code using the spread operator to solve both scenarios.
# const classA = ["Rahul", "Priya", "Amit"];
# const classB = ["Sneha", "Vikram", "Pooja"];

- Merge two arrays of student names
```
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

const allStudents = [...classA, ...classB];

console.log(allStudents);
// Output: ["Rahul", "Priya", "Amit", "Sneha", "Vikram", "Pooja"]
```

- Merge two objects: basic info + marks
```
const basicInfo = {
  name: "Arjun",
  age: 20
};

const marksInfo = {
  math: 85,
  science: 90
};
```
- Using the spread operator to combine both objects:
```
const studentDetails = {
  ...basicInfo,
  ...marksInfo
};

console.log(studentDetails);

```
